```python
html_doc=response.text
print(html_doc)
```

```python
soup=BeautifulSoup(html_doc,"html.parser")
```

```python
tables=soup.find_all("table")
print(len(tables))
```

```python
table=soup.find("table", class_="wikitable")
print(table)
```

```python
rows=table.find_all("tr")
print(len(rows))
```

```python
for row in rows:
    cols=row.find_all("td")
    print(len(cols))
```

```python
for row in rows:
    cols=row.find_all("td")
    if len(cols)>0:
        name_col=cols[0]
        print(name_col.text)
```

```python
countries=[]
for row in rows:
    cols=row.find_all("td")
    if len(cols)>0:
        name_col=cols[0]
        name_str:str=name_col.text
        name=name_str.split("[")[0].strip()
        print(name)
```

```python
countries=[]
for row in rows:
    cols=row.find_all("td")
    if len(cols)>0:
        name_col=cols[0]
        name_str:str=name_col.text
        name=name_str.split("[")[0].strip()
        countries.append(name)
print(countries) 

with open("countries.txt", "w")as file:
    for country in countries:
        file.write(country)
        file.write("\n")
```

```python
#QUEREMOS SACAR LA SEGUNDA COLUMNA.
country_dicts=[]
for row in rows:
    cols=row.find_all("td")
    if len(cols)>0:
        name_col=cols[0]
        name_str:str=name_col.text
        name=name_str.split("[")[0].strip()
        date_joined=cols[1].text.strip()
        print(date_joined)
        country_dict={
            "Name":name,
            "Date Joined":date_joined
            }
        country_dicts.append(country_dict)
        print(country_dicts)

with open("countries.csv", "w") as file:
    writer=DictWriter(file,fieldnames=["Name","Date Joined"])
    writer.writeheader()
    writer.writerows(country_dicts)
```
